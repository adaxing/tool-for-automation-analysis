```python
import os
import pandas as pd
import numpy as np

df1 = pd.read_excel('总名单.xlsx', encoding='utf-8', skiprows=1)
df2 = df1.iloc[:, 1:23]
df3 = pd.read_excel('新名单.xlsx', encoding='utf-8')
df2
```

```python
def find(x):
    result = np.where(df2 == x)
    try:
        return df2.columns[result[1][0]]
    except:
        return '未分组'
df3['备注'] = df3['最新名单'].apply(find)
df3
```

```python
result_lst = []
for index, i in enumerate(df2.columns):
    results = df3.iloc[np.where(df3['备注']==i)[0].tolist() ,0]
    results = results.reset_index(drop=True)
    result_lst.append(results)
df_final = pd.concat(result_lst, axis=1)
df_final.columns = df2.columns
df_final.to_excel(f'整理后表格.xlsx', encoding='utf-8', index=False, header=True)
```

```python
from openpyxl import load_workbook
from openpyxl.styles import Alignment
from openpyxl.styles import Side, Border
from openpyxl.styles import Font, PatternFill
import re
wb = load_workbook('整理后表格.xlsx')
ws = wb.active
font = Font(name='Palatino', size=18, bold=True)
alignment = Alignment(horizontal='center', vertical='center')
side = Side(style='none', color='000000')
border = Border(left=side, right=side, top=side, bottom=side)
bg_color = PatternFill(bgColor='778899', fill_type='lightGray')
for row in ws[f'{ws.dimensions}']:
    for cell in row:
        cell.alignment = alignment
        cell.border = border
ws.row_dimensions[1].height = 50
req = ':(\w)'
weight = re.findall(req, ws.dimensions)[0]
for cell in ws[1]:
    cell.fill = bg_color

wb.save('style.xlsx')
```
